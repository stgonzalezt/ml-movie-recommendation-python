```python
# notebooks/exploratory_data_analysis.ipynb
"""
Exploratory Data Analysis (EDA) de Recomendación de Películas

Descripción:
Este notebook realiza un análisis exploratorio de datos (EDA) para entender mejor la distribución y características de los datos procesados 
de películas y calificaciones.

Dependencias:
- pandas
- matplotlib
- seaborn

Autor:
stgonzalezt

Versión:
1.0
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos procesados
movies = pd.read_csv("../data/processed/movies_cleaned.csv")
ratings = pd.read_csv("../data/processed/ratings_cleaned.csv")
tags = pd.read_csv("../data/processed/tags_cleaned.csv")

# Mostrar las primeras filas de los datasets
display(movies.head())
display(ratings.head())
display(tags.head())

# Análisis exploratorio
sns.countplot(x='rating', data=ratings)
plt.title('Distribución de Calificaciones')
plt.show()
```
